## A Strategic Approach to a Ten Stock Weighted Investment
Using logitudinal data collected for each individual stock composing the *S&P 500*, calculations will be computed to determine the 50 best preforming stocks in the index.  These 50 stocks are then further analyzed to narrow the focus down to a weighted recommendation for a ten stock investment.  Users can then **paper invest** different investment amounts to predict their return over time.  The compuations used in this process consist of: *Sharpe Ratio, Monte Carlo Simulation, Moving Averages, Risk/Return Ratio, Relative Strength Index,* & *Correlation-Matrix.*

## Functionality of Sharpe Ratios 
<br>
Generally speaking, the greater the value of the Sharpe ratio, the stronger the risk-adjusted return. The Sharpe ratio can be calculated using data from the previous perfomance of the investment or expected values for the future of the investment. For this project, stocks with a Sharpe ratio of less than 1 are considered to be risky investment while stocks with a Sharpe ratio of more than 1 are considered strong investments because the investor recieves more return than risk. Stocks become more and more attractive as their Sharpe ratio increases from 1 to infinity.

## Functionality of Monte Carlo Simulations 
<br>
Monte Carlo simulations are used to predict possible outcomes when random variables are present. The forecasting model provides an explaination for the impact of the risk and uncertainty. The results are generated by assigning multiple values to random variables then averaging the outcomes all the different inputs provided.

## Functionality of Moving Averages
<br>
Simple moving averages smooth out a stock's price over a period of time by calculating a constantly updating average price over a set time period. In our case we have choosen to collect data using two time windows, one being a **50 day average** and the other  being a **200 day average**. Compared to analyzing just the past closing prices of a stock, using SMAs allows us to mitegate the impacts of short term random price fluctuations.    

## Functionality of Risk/Return Ratio
<br>
Generally, investors associate low levels of risk with low levels of return while high levels of risk are associated with high levels of return. Therefore, the risk/return tradeoff states that the higher the risk, the higher the potential reward. Graphically, the most appealing investments will be shown in the top left area, meaning the potential return outweighs the risk.  

## Functionality of Relative Strength Index
<br>
The relative strength index (RSI) is a momentum oscillator that provides insight about bullish and bearish price momentum movements. Typically, when the RSI is higher than 70% the assest is overbought, and when the RSI is lower than 30% the asset is oversold. 

## Functionality of Correlation-Matrix
<br>
Correlation is a statistical measurement that shows the degree to which two investment's price move in relation to each other. The correlation will be shown through a value in the range of -1.0 and +1.0 where the absoulte value is the degree of correlation and the sign of the number shows the direction of the correlation. The correlation is useful in picking stocks for our portfolio because it allows us to make sure that our investments are diverse across industries. Doing this is another way to protect our level of risk by making sure our portfolio' profit isn't dependant on the performance of one industry.
